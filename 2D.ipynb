```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('au.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

k1 = np.array([0])
k2 = np.array([[0.0625,0.0625,0.0625],[0.0625,0.0625,0.0625],[0.0625,0.0625,0.0625]])

k1output = cv2.filter2D(img,-1,k1)
k2output = cv2.filter2D(img,-1,k2)
plt.subplot(1,3,1)
plt.imshow(img)
plt.subplot(1,3,2)
plt.imshow(k1output)
plt.subplot(1,3,3)
plt.imshow(k2output)
```
